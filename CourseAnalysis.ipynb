```python
%pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util
import torch

model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Encode all course descriptions
yu_embeddings = model.encode(list(Yu.values()), convert_to_tensor=True)
uofv_embeddings = model.encode(list(uOfV.values()), convert_to_tensor=True)

# Calculate cosine similarity between all pairs of embeddings
cosine_scores = util.cos_sim(uofv_embeddings, yu_embeddings) #Note the change here.

# Calculate the overall similarity score
total_max_similarity = 0
for i in range(len(uOfV)):
    max_similarity = -1
    for j in range(len(Yu)):
        if cosine_scores[i][j] > max_similarity:
            max_similarity = cosine_scores[i][j].item()
    total_max_similarity += max_similarity

overall_similarity = total_max_similarity / len(uOfV)

# Print the overall similarity score
print(f"Overall Similarity (uOfV to Yu): {overall_similarity}")

# Print the most similar Yu course for each uOfV course
for i in range(len(uOfV)):
    max_similarity = -1
    most_similar_index = -1
    for j in range(len(Yu)):
        if cosine_scores[i][j] > max_similarity:
            max_similarity = cosine_scores[i][j].item()
            most_similar_index = j
    print(f"uOfV[{i}] is most similar to Yu[{most_similar_index}] with similarity: {max_similarity}")

# Example creating a similarity matrix
similarity_matrix = [[cosine_scores[i][j].item() for j in range(len(Yu))] for i in range(len(uOfV))]

print(similarity_matrix)
```
